### Importando as bibliotecas necessárias


```python
import numpy as np
import matplotlib.pyplot as plt
```

### Ajustando o modelo

```python
# 1. Dados simulados: tamanho da casa (m²) e preço (mil R$)
X = np.array([50, 60, 70, 80, 90, 100])
y = np.array([150, 180, 210, 240, 270, 300])

# 2. Normalizar X
X_mean = X.mean()
X_std = X.std()
X_norm = (X - X_mean) / X_std

# 3. Inicializar parâmetros
m = 0  # coeficiente angular
b = 0  # intercepto
learning_rate = 0.01
epochs = 1000
n = len(X_norm)

# 4. Gradiente descendente
for _ in range(epochs):
    y_pred = m * X_norm + b
    error = y_pred - y

    dm = (2/n) * np.dot(error, X_norm)
    db = (2/n) * np.sum(error)

    m -= learning_rate * dm
    b -= learning_rate * db

# 5. Fazer previsões finais
y_final = m * X_norm + b

```

### plotando os resultados

```python
# 6. Visualizar
plt.figure(figsize=(8, 5))
plt.scatter(X_norm, y, color='blue', label='Dados reais')
plt.plot(X_norm, y_final, color='red', label='Regressão Linear')
plt.xlabel("Tamanho da casa (normalizado)")
plt.ylabel("Preço (mil R$)")
plt.title("Regressão Linear com Gradiente Descendente")
plt.legend()
plt.grid(True)
plt.show()
```
